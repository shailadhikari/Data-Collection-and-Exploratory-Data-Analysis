```r
###############  Part-1 - 1  ####################

sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
sales2<-rpois(12,34)
#sales2
par(bg="#C885F0")
plot(sales1, col="white", type="o", ylim=c(0,100), xlab="Month", ylab="Sales")
title(main="Sales by Month")
lines(sales2, type="o", pch=5, lty=4, col="gray")
grid(nx=NULL, ny=NULL)
legend("topright", inset=.03, c("Sales1","Sales2"), fill=c("white","gray"), horiz=TRUE)
```

```r
###############  Part-1 - 2  ####################
sales<-read.table("salesdata.txt", header=T)
#sales
par(bg="gray")
barplot(as.matrix(sales), main="Sales Data", ylab= "Total",beside=T, col=rainbow(6))
```

```r
###############  Part-1 - 3  ####################
par(bg="#DEF5F7")
fn<-boxplot(sales,col=c("#A2CFF9","#A2F9F1"))$stats
text(1.34, fn[3,2], paste("Median = ", fn[3,2]), adj=0, cex=.7)
text(1.41, fn[3,1],paste("Median = ", fn[3,1]), adj=0, cex=.7)
grid(nx=NA, ny=NULL)

## Interpretation : Sales1 has larger range than Sales2. Sales1 median is significantly lower than Sales2.
```

```r
################ Part-1 - 4 ###############

fb1<-read.csv("FB.csv")
aapl1<-read.csv("AAPL.csv")
par(bg="#9B9C9C")
plot(aapl1$Adj.Close, col="#E1E4E5", type="o", ylim=c(150,200), xlab="Days", ylab="Price" )
lines(fb1$Adj.Close, type="o", pch=22, lty=2, col="#40B4FA")
legend("topright", inset=.05, c("Apple","Facebook"), fill=c("#E1E4E5","#40B4FA"), horiz=TRUE)
hist(main="Histogram of Apples's Adjusted closing price", xlab = "Adjusted Closing Price", 
     ylim=c(0,20), aapl1$Adj.Close, col=rainbow(8))

## Analysis : The adjusted closing price of facebook's stock is significantly and consistently higher than 
#             the adjusted closing price of apple's stock, except couple of instances. 
#             Also, the adjusted closing price of Apple's stock is mostly observed between the values 170 and 176.
```
